```python
# Importing the necessary libraries
import os
from docx import Document
import re

# Displaying the Folder Path
folder_path = "D:\\total"

# Function to extract years of work experience from a text using regular expressions
def extract_years_of_experience(text):
    pattern = r"(\d+)\s*(?:year[s]?)"
    match = re.search(pattern, text, re.IGNORECASE)
    if match:
        return int(match.group(1))
    else:
        return None

# Iterate through all Word documents in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".docx"):
        file_path = os.path.join(folder_path, filename)
        
        # Load the Word document
        doc = Document(file_path)
        
        # Extract the text from the document
        text = "\n".join([paragraph.text for paragraph in doc.paragraphs])
        
        # Extract the years of work experience using regular expressions
        years_of_experience = extract_years_of_experience(text)
        
        if years_of_experience:
            print(f"Years of work experience in {filename}: {years_of_experience}")
```
