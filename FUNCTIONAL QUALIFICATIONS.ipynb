```python
# importing necessary libraries

import glob
import os
import shutil
from docx import Document
from openpyxl import Workbook

def match_specializations(text):
    matched_specializations = []
    for specialization in post_graduation_spesilization:
        if specialization.lower() in text.lower():
            matched_specializations.append(specialization)
    return matched_specializations


def match_master_degrees(text):
    matched_degrees = []
    for abbreviation, degree in master_abbreviations_lower.items():
        if abbreviation in text.lower():
            matched_degrees.append(degree)
    return matched_degrees

def match_specializations1(text):
    matched_specializations1 = []
    for specialization in bechols_spesilization:
        if specialization.lower() in text.lower():
            matched_specializations1.append(specialization)
    return matched_specializations1

def match_bachelor_degrees(text):
    matched_degrees1 = []
    for abbreviation, degree in bachelor_abbreviations.items():
        if abbreviation in text.lower():
            matched_degrees1.append(degree)
    return matched_degrees1

# Set the source folder path where the CV files are located
source_folder = "D:\\total"

# Set the destination folder path where you want to save the extracted CVs and data
destination_folder = "D:\\"

# Create the destination folder if it doesn't exist
os.makedirs(destination_folder, exist_ok=True)

# Use glob to get a list of all CV files in the source folder
cv_files = glob.glob(os.path.join(source_folder, "*.pdf")) + glob.glob(os.path.join(source_folder, "*.docx"))

# Create a new Excel workbook
workbook = Workbook()
worksheet = workbook.active

# Write headers to the worksheet
worksheet.append(["CV File","Specialization", "Qualification","Bacholers Specialization","Bachlores Qualifications"])

# Iterate over each CV file
for i, cv_file in enumerate(cv_files, start=1):
    # Skip temporary Word files
    if os.path.basename(cv_file).startswith("~$"):
        print(f"Skipping file {cv_file} - temporary Word file.")
        continue

    # Read the content of the CV file
    _, ext = os.path.splitext(cv_file)
    if ext == '.pdf':
        with open(cv_file, 'rb') as file:
            cv_content = file.read().decode('utf-8', errors='ignore')
    elif ext == '.docx':
        doc = Document(cv_file)
        paragraphs = [p.text for p in doc.paragraphs]
        cv_content = '\n'.join(paragraphs)
    else:
        print(f"Skipping file {cv_file} - unsupported file format.")
        continue

    # Extract the qualifications and bachelor degrees from the CV content
    matched_specializations = match_specializations(cv_content)
    matched_degrees = match_master_degrees(cv_content)
    matched_specializations1 = match_specializations1(cv_content)
    matched_degrees1 = match_bachelor_degrees(cv_content)

    # Extract only the CV name from the file name
    cv_name = os.path.basename(cv_file)

    # Create a new filename for the extracted CV based on the location
    filename = f"cv{i}={cv_name}.pdf"

    # Set the destination path for the extracted CV
    destination_path = os.path.join(destination_folder, filename)

    # Copy the CV file to the destination folder with the new filename
    shutil.copyfile(cv_file, destination_path)

    # Write the CV file name, qualifications, and bachelor degrees to the worksheet
    worksheet.append([cv_name, ', '.join(matched_specializations), ', '.join(matched_degrees),','.join(matched_specializations1),','.join(matched_degrees1)])

# Save the workbook as an Excel file
excel_file = os.path.join(destination_folder, "cv_function.xlsx")
workbook.save(excel_file)

# Print the path to the Excel file
print(f"CV data saved to: {excel_file}")
```
