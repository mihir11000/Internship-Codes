```python
import PyPDF2

def pdf_to_text(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)  # Use PdfFileReader instead of PdfReader
        num_pages = len(reader.pages)  # Get the number of pages using numPages

        text = ''
        for page in range(num_pages):
            text += reader.pages[page].extract_text()

        return text

# Example usage
pdf_path = 'D:\\first\\Biswaranjan kar  - Branch manager - 10 Yrs 0 Month (1).pdf'
text = pdf_to_text(pdf_path)
print(text)
```

```python
from nltk import sent_tokenize, word_tokenize, pos_tag

def extract_years_of_experience(text):
    sentences = sent_tokenize(text)
    years_of_experience = []

    for sentence in sentences:
        words = word_tokenize(sentence)
        tagged_words = pos_tag(words)

        for i in range(len(tagged_words) - 1):
            if tagged_words[i][1] == 'CD' and tagged_words[i + 1][0] == 'years':
                years_of_experience.append(int(tagged_words[i][0]))

    return years_of_experience

years_of_experience = extract_years_of_experience(text)
print(years_of_experience)
```

```python
def match_designations(text, designations):
    pattern = r'\b(?:{})\b'.format('|'.join(map(re.escape, designations)))
    matches = re.findall(pattern, text, flags=re.IGNORECASE)
    return matches


designations = ['Branch Manager',
                'Manager',
                'Deputy Manager',
                'Senior  Officer',
                'Deputy  Branch  Manager',
                'Assistant  manager',
                'Customer  Service  Officer',
                'Assistant Branch Manager',
                'Sales Manager','Area Officer',
                'Desk Officer'
                'Desk Manager',
                'Sales  Executive',
                'Area Manager',
                'Branch Operations Manager',
                'Customer  Service  manager',
                'Customer Respresentative Manager']

matches = match_designations(text, designations)
print(matches)
```

```python
import spacy

def match_locations(text):
    # Load the English language model in spaCy
    nlp = spacy.load('en_core_web_sm')
    
    # Process the text
    doc = nlp(text)
    
    # Initialize a set to store matched locations
    matched_locations = set()
    
    # Extract entities from the processed text
    for entity in doc.ents:
        # If the entity label is 'GPE' (geopolitical entity) or 'LOC' (location)
        if entity.label_ in ['GPE', 'LOC']:
            matched_locations.add(entity.text)
    
    return matched_locations

# Match locations from the text
matched_locations = match_locations(text)

# Print the matched locations
print("Matched locations:")
for location in matched_locations:
    print(location)
```
